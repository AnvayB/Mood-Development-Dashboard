```python
import pandas as pd
monthly_hi = pd.read_csv("./data/mood_monthly_hi.csv")
```

```python
second_min_hi_row = monthly_hi.nsmallest(2, 'happiness_index').iloc[1]
max_hi_row = monthly_hi.loc[monthly_hi['happiness_index'].idxmax()]

print(f"Min happiness_index: {second_min_hi_row['happiness_index']} (Year: {second_min_hi_row['year']}, Month: {second_min_hi_row['month']})")
print(f"Max happiness_index: {max_hi_row['happiness_index']} (Year: {max_hi_row['year']}, Month: {max_hi_row['month']})")
```

```python
sorted_hi = monthly_hi['happiness_index'].sort_values()
print(sorted_hi.to_list())
```

```python
import numpy as np

# Divide the sorted_hi into 3 roughly equal parts
sorted_hi_list = sorted_hi.to_list()
partitions = np.array_split(sorted_hi_list, 3)
part1, part2, part3 = partitions[0], partitions[1], partitions[2]

print("Part 1:", part1.tolist())
print("Part 2:", part2.tolist())
print("Part 3:", part3.tolist())
```
